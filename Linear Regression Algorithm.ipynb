```python
print("Hello World")

import numpy as np
import matplotlib.pyplot as plt

def makeWeightArray(X,Xtran,y):
    prein = np.matmul(Xtran,X)
    postin = np.linalg.pinv(prein)
    prey = np.matmul(postin,Xtran)
    weight = np.matmul(prey,y)
    return weight

def getRows(matrix):
    return (matrix.shape[0])

def getColumns(matrix):
    return (matrix.shape[1])

def printWeight(weight):
    for i in range(0,getRows(weight),1):
        print("W"+str(i)+" = "+str(weight[i]))
        
def printRealEstateWeight(weight):
    print("Base Weight: "+str(weight[0]))
    print("Bedroom's Weight: "+str(weight[1]))
    print("Bathroom's Weight: "+str(weight[2]))
    print("Square Foot's Weight: "+str(weight[3]))
    print("Lot Square Foot's Weight: "+str(weight[4]))
    print("Year's Weight: "+str(weight[5]))
    

def examineRealEstateDelta(weight, X,output):
    rows = getRows(X)
    delta = np.zeros(rows)
    
    for i in range(0,rows,1):
        addition = 0;
        for j in range (0,6,1):
            addition =  addition + X[i][j]*weight[j]
            
        addition = abs(addition - output[i])
        delta[i] = addition
        
    return delta

def calculateTotalDelta(delta):
    total = 0;
    for i in range (0,getRows(delta),1):
        total = total +delta[i]
    return float(total)
    
def computePriceRealEstate(weight):
    print("Ok, How many bedroom's does the house have?")
    bed = float(input("Input number of bedrooms: "))
    print("How many Bathrooms?")
    bath = float(input("Input number of bathrooms: "))
    print("Give me the Sqft of the house.")
    sqft = float(input("The sqft of the house: "))
    print("How about the sqft of the lot?")
    sqftlot = float(input("Sqft of the lot: "))
    print("In which year was the house completed?: ")
    year = float(input("The year: "))
    
    final = weight[0] + weight[1]*bed + weight[2]*bath 
    
    final = final + weight[3]*sqft + weight[4]*sqftlot + weight[5]*year
    return final
    
```
